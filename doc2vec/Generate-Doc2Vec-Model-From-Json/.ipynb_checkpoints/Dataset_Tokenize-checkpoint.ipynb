# Tokenize the notebook source code

```python
import numpy as np
import pandas as pd
from tokenize import tokenize
from io import BytesIO
```

```python
df = pd.read_json("../data/all-notebooks.json",orient="index")
```

```python
df.head()
```

```python
def tokenize_code(code, cell_type):
    # if markdown or raw, split with " "
    if cell_type != "code":
        return []
    tokenized_code = []
    tokens = []
    try:
        tokens = tokenize(BytesIO(code.encode('utf-8')).readline)
    except IndentationError:
        return []
    
    for tok in tokens:
        ret = ""
        # type 4 is NEWLINE
        if (tok.type == 4 or tok.type == 61):
            ret = "[NEWLINE]"
        # type 5 is INDENT
        elif (tok.type == 5):
            ret = "[INDENT]"
        else:
            ret = tok.string
#         print(tok)
#         print(f"Type: {tok.exact_type}\nString: {tok.string}\nStart: {tok.start}\nEnd: {tok.end}\nLine: {tok.line.strip()}\n======\n")
        tokenized_code.append(ret)
    return tokenized_code
```

```python
subdf = df.head(100)
```

```python
subdf['tokenized_source'] = df.apply(lambda row: tokenize_code(row['source'], row['cell_type']), axis=1)
```
